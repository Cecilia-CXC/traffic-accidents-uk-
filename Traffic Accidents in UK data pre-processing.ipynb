```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Road Accident Data.csv')
```

```python
df['Accident Date'] = pd.to_datetime(df['Accident Date'], format='%d-%m-%Y')
df.info()
```

```python
#For missing values under Time column remove them
df.dropna(subset=['Time'], inplace=True)
```

```python
#For missin values under Carriageway_Hazards replace with 'Unknown'
df['Carriageway_Hazards'].fillna('Unknown', inplace=True)
```

```python
#check for duplicates
df.duplicated().sum()
```

```python
for column in df.select_dtypes(include='object').columns:
    print(f"Unique values for column '{column}':")
    print(df[column].unique())
    print("\n" + "-"*50 + "\n")
```

```python
#Typo fixing
df['Junction_Control'] = df['Junction_Control'].replace({
    'Auto traffic sigl': 'Auto traffic signal'
})

df['Accident_Severity'] = df['Accident_Severity'].replace({
    'Fetal': 'Fatal'
})

# Fix Local Authority typos
df['Local_Authority_(District)'] = df['Local_Authority_(District)'].replace({
    'Crewe and ntwich': 'Crewe and Nantwich',
    'North Larkshire': 'North Lanarkshire',
    'South Larkshire': 'South Lanarkshire',
    'Stevege': 'Stevenage'
})
```

```python
# Vehicle type: merge rare types into 'Other'
rare_vehicles = ['Agricultural vehicle', 'Ridden horse', 'Other vehicle']
df['Vehicle_Type_Simple'] = df['Vehicle_Type'].replace({v: 'Other' for v in rare_vehicles})
```

```python
# Vehicle type: merge rare types into 'Other'
rare_vehicles = ['Agricultural vehicle', 'Ridden horse', 'Other vehicle']
df['Vehicle_Type_Simple'] = df['Vehicle_Type'].replace({v: 'Other' for v in rare_vehicles})
```

```python
import datetime

def get_time_of_day(time_obj):
    if isinstance(time_obj, datetime.time):
        hour = time_obj.hour
    else:
        # Handle cases where the 'Time' column might still be a string (e.g., '14:30')
        try:
            hour = pd.to_datetime(time_obj, format='%H:%M').hour
        except ValueError:
            return 'Unknown'

    if 0 <= hour <= 5:
        return 'Early Morning'
    elif 6 <= hour <= 11:
        return 'Morning'
    elif 12 <= hour <= 16:
        return 'Afternoon/Peak Hour'
    elif 17 <= hour <= 20:
        return 'Evening/Rush Hour'
    else:
        return 'Night'

# Ensure the 'Time' column is in a proper time format first if not already
# The previous step was to ensure this conversion, but re-checking here for robustness
if not pd.api.types.is_datetime64_any_dtype(df['Time']) and not isinstance(df['Time'].iloc[0], datetime.time):
    df['Time'] = pd.to_datetime(df['Time'], format='%H:%M').dt.time

df['Time_of_Day'] = df['Time'].apply(get_time_of_day)

print(df[['Time', 'Time_of_Day']].head())
```

```python
#Dropping of columns that are not useful
keep_cols = [
    "Accident Date",
    "Day_of_Week",
    "Latitude",
    "Longitude",
    "Accident_Severity",
    "Number_of_Casualties",
    "Number_of_Vehicles",
    "Speed_limit",
    "Road_Type",
    "Road_Surface_Conditions",
    "Junction_Control",
    "Junction_Detail",
    "Light_Conditions",
    "Weather_Conditions",
    "Urban_or_Rural_Area",
    "Carriageway_Hazards",
    "Vehicle_Type_Simple",
    "Time_of_Day",
    "Accident_Index"
]

# Keep only those that exist
df = df[[c for c in keep_cols if c in df.columns]]

print("Final columns kept:", df.columns.tolist())
```
